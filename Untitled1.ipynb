```python
from bs4 import BeautifulSoup

page_count = 1
list_life = []

while True:
    try:
        with open(f"Scraped/quotes{page_count}.html", "r", encoding="utf-8") as f:
            content = f.read()

        soup = BeautifulSoup(content, "lxml")
        all_quotes = soup.select("div.quote")

        for q in all_quotes:
            all_tags = []
            for t in q.select(".tags .tag"):
                all_tags.append(t.get_text())

            if "life" in all_tags:
                list_life.append({
                    "quote": q.select_one("span.text").get_text(),
                    "author": q.select_one("small.author").get_text()
                })

        page_count += 1

    except FileNotFoundError:
        print("All files processed. Stopping loop.")
        break
```

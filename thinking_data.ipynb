```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store_data.json")
```

```python
print(type(data))
```

```python
def cleandata(data):
    text_to_num = {"one": 1,"two":2,"three":3,"four":4,"five": 5}
    cleaned_data=[]
    unique_users=set()
    for user in data:
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating= text_to_num[raw_rating]
        user["rating"] = raw_rating

        raw_age = user.get("age")
        if(raw_age == None):
            user["age"]=None

        if(user["name"].strip() in unique_users):
            continue
        else:
            unique_users.add(user["name"].strip())

        cleaned_data.append(user)

    return cleaned_data
```

```python
data = cleandata(data)
```

```python
def insight(data):
    total_rating=0
    poor_rating=0
    for user in data:
        total_rating += float(user["rating"])
        if(float(user["rating"]) < 3):
            poor_rating += 1
    print(f"total_rating = {total_rating/len(data)}")
    print(f"poor_rating = {poor_rating/len(data) * 100}")
```

```python
insight(data)
```

```python
def recom(data):
    recommended=[]
    for user in data:
        curr_recom={}
        curr_recom["name"]=user["name"]
        if (float(user["rating"])>=4):
            curr_recom["brand"]="Apple"
        else:
            curr_recom["brand"]="Samsung"
        recommended.append(curr_recom)

    return recommended
```

```python
recom(data)
```

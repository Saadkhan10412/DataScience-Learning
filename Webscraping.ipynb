```python
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
```

```python
print(res.status_code)
```

```python
if res.status_code == 200:
    print(res.headers)
with open("Scraped/data1.html", "w", encoding="utf8") as f:
        f.write(res.text)
```

```python
from bs4 import BeautifulSoup 
with open("Scraped/data1.html","r") as f:
    html_content=f.read()

soup=BeautifulSoup(html_content,"lxml")
```

```python
all_h3=soup.find_all("h3")
all_countries=[]
for h3 in all_h3:
    name=h3.get_text(strip=True)
    population=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(all_countries,columns=["Name","population"])
```

```python
df
```

```python
df.to_csv("Scraped/datalover.csv",index=False)
```
